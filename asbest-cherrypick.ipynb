```python
import matplotlib.pyplot as plt
import pandas as pd

air_pollution_deaths = 6_700_000   
fart_share_estimate = 0.045       
fart_related_deaths = int(air_pollution_deaths * fart_share_estimate)

df = pd.DataFrame({
    "Cause": ["Air Pollution (All)", "Farting*", "Asbestos"],
    "Deaths": [air_pollution_deaths, fart_related_deaths, 255_000]
})

plt.figure(figsize=(8,6))
bars = plt.bar(df["Cause"], df["Deaths"], color=["#a6cee3","#b2df8a","#fb9a99"])

#  Keep full numbers on top of bars
for b in bars:
    plt.text(b.get_x() + b.get_width()/2, b.get_height() + 50_000,
             f"{int(b.get_height()):,}", ha='center', va='bottom')

plt.title("Farting: A Hidden Contributor to Air Pollution Deaths?")
plt.ylabel("Estimated Deaths (Millions per Year)")

plt.ticklabel_format(style='plain', axis='y')
plt.yticks(
    ticks=[i * 1_000_000 for i in range(0, 8)],  
    labels=[str(i) for i in range(0, 8)]          
)

plt.ylim(0, 7_000_000)
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.figtext(
    -0.02, -0.02,
    "*Fart share estimated as 4.5% of global methane emissions applied to air-pollution deaths.\n"
    "Air pollution deaths from Our World in Data (Ritchie & Roser 2023). Asbestos deaths from Furuya et al. 2018.",
    fontsize=8
)
plt.tight_layout()
plt.show()
```
